```python
import pandas as pd

file_path = r'/content/drive/MyDrive/Retail_Transactions_Dataset_M.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())

# Number of samples (rows)
num_samples = len(df)

# Number of unique items in another specified column, e.g., 'Product'
# Replace 'Product' with the column you want to analyze
num_unique_items = df['Product'].nunique()

print(f"Number of samples: {num_samples}")
print(f"Number of unique items in 'Product': {num_unique_items}")
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each category in the 'Payment_Method' column
payment_counts = df['Payment_Method'].value_counts()

# Plot a bar chart
plt.figure(figsize=(8, 6))
payment_counts.plot(kind='bar', color=['blue', 'green', 'red', 'purple'])
plt.title('Distribution of Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
